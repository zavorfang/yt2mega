```python
from mega import Mega
import os, datetime, time

mega = Mega()
usernames = ""
passwords = ""
m = mega.login(usernames[1], passwords[1])

def main():
  # Initializing start time
  begin = time.time()

  # Showing all uploadable directories
  cwd = os.listdir(os.getcwd())
  print(cwd)
  # Selecting local directory to upload
  num = input("Select directory to download: ")
  if num.isdigit() and int(num) <= 4:
    num = int(num)
  path = cwd[num]
  # Checking if prior operation was engaged
  folder = m.find(path)

  if folder is None:
    print(path, " doesn't exist")
    # Creating online(mega) target folder
    m.create_folder(path)
    # Selecting online(mega) target folder just created hitherto
    folder = m.find(path)

  # Initializing total filesize
  filesize = 0
  # Looping through all files in current working directory
  for file in os.listdir(path):
    # Checks if file already exists before uploading
    exists = m.find(file)
    if exists is None:
      # Showing file name
      print("Uploading", file)

      # Showing appoximated file size through tenary
      size = int(os.path.getsize(os.path.join(os.getcwd(), path, file)))
      print("Size:", f"{round(size/(1024*1024))} MB" if size > 1024*1024 else f"{round(size/1024)} KB" if size > 1024 else f"{size} Bytes")
      
      # Uploading to mega and getting link
      print("LINK: ", dl_mega(path, file, folder))
      filesize += size
    else:
      print(file, "already exists")
    
    # Inserting new line for readability
    print()

  # Detailing final details of transfer
  end = time.time()
  filesize = round(filesize/(1024*1024))

  print("Total size copied: ", filesize, " MB")
  print("Elapsed Time: ",int((end-begin)//60),"min :", int((end-begin)%60), "sec")
  print(datetime.datetime.now())

def dl_mega(path, file, folder):
  file = m.upload(f'{os.path.join(os.getcwd(), path, file)}', folder[0])
  return m.get_upload_link(file)

main()
```
